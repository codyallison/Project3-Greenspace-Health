```python
## Import Dependencies

import pandas as pd
import re
```

```python
## Create the Dataframe and list the column names

df = pd.read_csv('../Resources/CountyHealthData.csv')
list(df.columns)
```

```python
## Clean the DataFrame and only show the columns we need

df = df[['Year',
 'StateAbbr',
 'LocationName',
 'Data_Value_Type',
 'Data_Value',
 'TotalPopulation',
 'CategoryID',
 'MeasureId',
 'DataValueTypeID',
 'Short_Question_Text',
 'Geolocation',]]
df.dropna()
```

```python
## Fix the format of the Geolocation column

## Create an empty list
trimmed_geo = []
## Iterate through the rows
for i in df['Geolocation']:
    ## Set each item to a strin
    s = str(i)
    # get lon-lat from "POINT (lon lat)"
    loc = s[7:-1]
    ## Turn the returned string into a list
    l = loc.split(' ')
    ## Reverse the list
    l.reverse()
    ## Use a list comprehension to turn the string values in the list back in to floats
    if len(l) == 2:
    
        p = [float(x) for x in l]
    ## Return nothing if there is no value for the geolocation
    else:
        p = ''
    ## Add the float list to the empty list
    trimmed_geo.append(p)
        
## Set the Geolocation column equal to the list of lists
df['Geolocation'] = trimmed_geo
```

```python
## Show the DataFrame again with the new location values
df
```

```python
## Export it to a file

df.to_csv('../Resources/CleanedHealthCountyData.csv')
```
