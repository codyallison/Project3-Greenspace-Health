```python
## Import Dependencies

import pandas as pd
import re
```

```python
## Create the Dataframe and list the column names

df = pd.read_csv('../Resources/CountyHealthData.csv')
list(df.columns)
```

```python
list(df['Measure'].unique())
```

```python
## Clean the DataFrame and only show the columns we need

df = df[['Year',
 'StateAbbr',
 'LocationName',
 'Data_Value_Type',
 'Data_Value',
 'TotalPopulation',
 'CategoryID',
 'MeasureId',
 'DataValueTypeID',
 'Short_Question_Text',
 'Geolocation']]
df.dropna()
```

```python
new_df = df.loc[df['Measure'].isin(
                        ['High blood pressure among adults aged >=18 years'])]
new_df
```

```python
df['Measure'].isin(['High blood pressure among adults aged >=18 years'])
```

```python
list(df['Short_Question_Text'].unique())
```

```python
## Fix the format of the Geolocation column

## Create an empty list
lats = []
lons = []

## Iterate through the rows
for i in df['Geolocation']:
    ## Set each item to a string
    s = str(i)
    # get lon-lat from "POINT (lon lat)"
    loc = s[7:-1]
    ## Turn the returned string into a list
    l = loc.split(' ')
    ## Reverse the list
    
    try:
        lats.append(l[1])
        
    except:
        lats.append('')
        
    try:
        lons.append(l[0])
        
    except:
        lats.append('')
           
        
        
        
    ## Use a list comprehension to turn the string values in the list back in to floats
df['Lat'] = lats
df['Lon'] = lons

```

```python
## Show the DataFrame again with the new location values
df = df.drop(columns = ['CategoryID', 'MeasureId', 'Geolocation'])
df
```

```python
## Changing column types
df['Lat'] = pd.to_numeric(df['Lat'])
df['Lon'] = pd.to_numeric(df['Lon'])
```

```python
## Export it to a file

df.to_csv('../Resources/CleanedHealthCountyData.csv')
```
